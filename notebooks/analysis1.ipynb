# Method Chaining

```python
import pandas as pd
import numpy as np
#method chaining side
```

```python
    def first_python_module(url_or_path_to_csv_file):
    
       # Method Chain 1 (Load data,rename columns, get rid of null values and drop necessary data with missing data)\n",
    
        df1 = (
              pd.read_csv(url_or_path_to_csv_file)    
              .drop(['NBA_FLAG','JERSEY','PIE','TEAM_ABBREVIATION','TEAM_ID','ALL_STAR_APPEARANCES','GAMES_PLAYED_CURRENT_SEASON_FLAG','FROM_YEAR','TO_YEAR','DLEAGUE_FLAG','TEAM_CODE','TEAM_CITY','DISPLAY_FIRST_LAST','PLAYER_SLUG','FIRST_NAME','LAST_NAME','DISPLAY_FI_LAST','GAMES_PLAYED_FLAG','PLAYERCODE'], axis=1)
             .rename(columns={'DISPLAY_LAST_COMMA_FIRST': 'FULL NAME', 'PTS':'PPG', 'AST':'APG','REB':'RBG'})  
            .dropna(axis=0)
            )
        
        df2 = (
            df1.dropna(subset=['HEIGHT','PPG','APG','RBG'])
                .loc[lambda x: x['HEIGHT']>84]
                .assign(OverallScore= (df1['APG'] * 0.3)+ (df1['PPG'] * 0.4)+(df1['RBG'] * 0.3))
                .reset_index(drop = True) 
        )
    
        return df2
```

```python
 pko = first_python_module("../data/raw/Player_Attributes.csv")
```

 I just tried and tested my method. I added a new column called overall score which takes into account a players points per game, assists per game and rebounds per game and aggregates together. However, points do have higher impact on the aggreagetion while assists and rebounds are considered the same. I also filtered the height to only players above 84 inches because the dataset had up to 3500 rows.
